```python
f = open("content.txt", "r", encoding="utf-8")
s = f.read()

f.close()
s
```

```python
listStr = s.split()
wcText = ''
#특수문자 삭제
for listWord in listStr:
    word = listWord.strip("'\"‘’“”[]")
    wcText = wcText + " " + word
wcText
```
